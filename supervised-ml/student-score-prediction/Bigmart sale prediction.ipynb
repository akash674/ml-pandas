```python
# importing all the tools which will be use in this project
import pandas as pd
from sklearn.linear_model import  LinearRegressionimport pandas as pd
from sklearn.linear_model import  LinearRegression
```

```python
#creating the data frame of pandas on training data  and load the data 
train_df=pd.read_csv("train.csv")
train_df.head()
```

```python
#checking the all the information about data
train_df.info()
```

```python
#checking the information of individual column
rain_df["Outlet_Size"].value_counts()
```

```python
#After the checking the information , filling the null values 
def fillna_null_value(train_df):
    train_df["Item_Weight"]=train_df["Item_Weight"].fillna(train_df["Item_Weight"].mean())
    train_df["Outlet_Size"]=train_df["Outlet_Size"].fillna("Medium")
    return train_df
```

```python
#checking the all null values filled or not 
train_df=fillna_null_value(train_df)
train_df.info()
```

```python
#deleting the unneccessary column(for increasing the performance )
train_df=train_df.drop(["Item_Identifier","Outlet_Identifier"],axis=1)
train_df.info()
```

```python
#mapping the value count column
def get_mapping(df,col_name):
    x=df[col_name].value_counts()
    l_value=list(x.index)
    mapping={}
    for i in range (0, len(l_value)):
        mapping[l_value[i]]=i
        
    return mapping    
```

```python
#changing the all string into interger and for chnages i am using here NumPy
def change_to_num(df):
    mapping=get_mapping(df,"Item_Fat_Content")
    df["Item_Fat_Content"]=df["Item_Fat_Content"].map(lambda x:mapping[x])
    
    mapping=get_mapping(df,"Item_Type")
    df["Item_Type"]=df["Item_Type"].map(lambda x:mapping[x])
    
    mapping=get_mapping(df,"Outlet_Size")
    df["Outlet_Size"]=df["Outlet_Size"].map(lambda x:mapping[x])
    
    mapping=get_mapping(df,"Outlet_Location_Type")
    df["Outlet_Location_Type"]=df["Outlet_Location_Type"].map(lambda x:mapping[x])
    
    mapping=get_mapping(df,"Outlet_Type")
    df["Outlet_Type"]=df["Outlet_Type"].map(lambda x:mapping[x])
    
    return df

      
```

```python
#so here all the data changed into interger
train_df=change_to_num(train_df)
train_df.head()
```

```python
#dropping one more column
y=train_df["Item_Outlet_Sales"]

x=train_df.drop(["Item_Outlet_Sales"],axis=1)
x.head()
```

```python
#checking the column information we dropped
y.head()
```

```python
#using the linear regression for training the data 
lr=LinearRegression()
```

```python
#using the fit function for training the data 
lr.fit(x,y)
```
