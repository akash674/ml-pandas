#Loan prediction using Ml and Python

```python
# importing the pandas and Logistic regression
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
# creating the dataframe of  the train data and load the data 
train_df=pd.read_csv("train.csv")
train_df.head()
```

```python
#displaying all information of the data 
train_df.info()
```

```python
#dropping the unneccessary columns(loan_ID and Gender)

train_df=train_df.drop(['Loan_ID','Gender'],axis=1)
train_df.head()
```

```python
# checking the null value
train_df['Married'].value_counts()
```

```python
#filling the null values with 0
train_df['Married']=train_df['Married'].fillna('0')
train_df.info()
```

```python
#filling the null values
train_df['Dependents']=train_df['Dependents'].fillna(0)
train_df.info()
```

```python
#filling the null values
train_df['Self_Employed']=train_df['Self_Employed'].fillna('No')
train_df.info()
```

```python
#filling the null values
train_df['LoanAmount']=train_df['LoanAmount'].fillna(5000)
train_df.info()
```

```python
#filling the null values
train_df['Loan_Amount_Term']=train_df['Loan_Amount_Term'].fillna(360.0)
train_df.info()
```

```python
#filling the null values
train_df['Credit_History']=train_df['Credit_History'].fillna(1.0)
train_df.info()
```

```python
#checking the loan status
y=train_df['Loan_Status']
y.head()
```

```python
# dropping the Loan Status column
x=train_df.drop(['Loan_Status'],axis=1)
x.head()
```

```python
#changing the all string into integer 
def change_to_int(df):
    df['Dependents']=df['Dependents'].map(lambda x: 3 if x=="3+" else x)
    df['Married']=df['Married'].map(lambda x: 1 if x=="Yes" else 0)
    df['Education']= df['Education'].map(lambda x: 1 if x=="Graduate" else 0)
    df['Self_Employed']= df['Self_Employed'].map(lambda x: 1 if x=="Yes" else 0)
    mapping={"Urban":1, "Rural":0,"Semiurban":2}
    df['Property_Area']= df['Property_Area'].map(lambda x: mapping[x])
    return df

    
```

```python
#checking the all columns are integer or not because we chnaged the string into integer 
train_df=change_to_int(x)
train_df.describe()
```
